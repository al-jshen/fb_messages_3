Below, we read in messages and create an LSTM network to mimic the message style of a user. 

```python
# Read file
with open('messages/inbox/GOTTAGOFASTER_dBGU9wOQpg/message_1.json', 'r') as f:
    m = json.load(f)
dat = pd.DataFrame(m['messages'])
```

```python
# Get list of names
names = list(set(dat['sender_name']))
names
```

```python
# Preprocessing
dat = dat[dat.sender_name != 'Ansel Hartanto']
dat = dat[dat.sender_name != 'Bob Bbob']
dat = dat.loc[:, ['content', 'sender_name']]
dat = dat[::-1].reset_index().iloc[:,1:]
```

```python
# Convert all user's messages to string
def get_user_text(user_idx):
    text = dat[dat.sender_name == names[user_idx]]['content'].to_string(header=False, index=False)
    text = ''.join([i.strip() + '\n' for i in text.splitlines()])
    return text
```

```python
# Encode text by mapping each char to a unique int
def encode(text):
    chars = tuple(set(text))
    int2char = dict(enumerate(chars))
    char2int = {ch: i for i, ch in int2char.items()}
    encoded = np.array([char2int[ch] for ch in text])
    return encoded
```

```python
def one_hot_encode(arr, n_labels):
    # initialize all 0s
    one_hot = np.zeros((np.multiply(*arr.shape), n_labels))
    # fill with 1s
    one_hot[np.arange(one_hot.shape[0]), arr.flatten()] = 1
    # reshape
    one_hot = one_hot.reshape((*arr.shape, n_labels))
    return one_hot
```

```python
def get_batches(arr, batch_size, seq_length):
    # trim text to get complete batches
    n_batches = len(arr) // (batch_size * seq_length)
    arr = arr[:n_batches * batch_size * seq_length]
    arr = arr.reshape((batch_size, -1))
    
    for n in range(0, arr.shape[1], seq_length):
        x = arr[:, n:n+seq_length]
        y = np.zeros_like(x)
        try:
            y[:, :-1], y[:, -1] = x[:, 1:], arr[:, n+seq_length]
        except IndexError:
            y[:, :-1], y[:, -1] = x[:, 1:], arr[:, 0]
        yield x, y
```

```python
train_on_gpu = torch.cuda.is_available()
```

```python
class CharRNN(nn.Module):
    
    def __init__(self, text, n_hidden=256, n_layers=2,
                               drop_prob=0.5, lr=0.001):
        super().__init__()
        self.drop_prob = drop_prob
        self.n_layers = n_layers
        self.n_hidden = n_hidden
        self.lr = lr
        
        self.chars = tuple(set(text))
        self.int2char = dict(enumerate(self.chars))
        self.char2int = {ch: ii for ii, ch in self.int2char.items()}
        self.n_chars = len(self.chars)
        
        self.lstm = nn.LSTM(self.n_chars, self.n_hidden, self.n_layers,
                               dropout=self.drop_prob, batch_first=True)
        self.drop = nn.Dropout(p=self.drop_prob)
        self.fc = nn.Linear(self.n_hidden, self.n_chars)
        
    def forward(self, x, hidden):
                
        out, hidden = self.lstm(x, hidden)
        
        out = self.drop(out)
        out = out.contiguous().view(-1, self.n_hidden)
        out = self.fc(out)
        
        return out, hidden
    
    def init_hidden(self, batch_size):

        weight = next(self.parameters()).data
        
        if (train_on_gpu):
            hidden = (weight.new(self.n_layers, batch_size, self.n_hidden).zero_().cuda(),
                  weight.new(self.n_layers, batch_size, self.n_hidden).zero_().cuda())
        else:
            hidden = (weight.new(self.n_layers, batch_size, self.n_hidden).zero_(),
                      weight.new(self.n_layers, batch_size, self.n_hidden).zero_())
        
        return hidden
```

```python
def train(net, dat, epochs=5, test_pct=0.2, lr=1e-3, batch_size=10, seq_length=50, print_every=10):
    net.train()
    
    optimizer = optim.Adam(net.parameters(), lr=lr)
    criterion = nn.CrossEntropyLoss()
    
    split_idx = int(len(dat) * (1-test_pct))
    train_dat, test_dat = dat[:split_idx], dat[split_idx:]
                    
    if train_on_gpu:
        net.cuda()
    
    counter = 0
    
    for e in range(epochs):
        h = net.init_hidden(batch_size)
        net.train()
        
        for x, y in get_batches(train_dat, batch_size, seq_length):
            counter += 1
            x = one_hot_encode(x, len(net.chars))
            features, labels = torch.from_numpy(x), torch.from_numpy(y)
            
            if train_on_gpu:
                features, labels = features.cuda(), labels.cuda()
                
            h = tuple([i.data for i in h])
            
            net.zero_grad()
            
            output, h = net(features.float(), h)
            
            loss = criterion(output, labels.view(batch_size * seq_length))
            loss.backward()
            optimizer.step()
            
            if counter % print_every == 0:
                net.eval()
                h = net.init_hidden(batch_size)
                test_losses = []
                
                for x, y in get_batches(test_dat, batch_size, seq_length):
                    x = one_hot_encode(x, len(net.chars))
                    features, labels = torch.from_numpy(x), torch.from_numpy(y)
                    if train_on_gpu:
                        features, labels = features.cuda(), labels.cuda()
                    h = tuple([i.data for i in h])
                    output, h = net(features.float(), h)
                    test_loss = criterion(output, labels.view(batch_size * seq_length))
                    test_losses.append(test_loss.item())
                                    
                print("Epoch: {}/{}...".format(e+1, epochs),
                      "Step: {}...".format(counter),
                      "Loss: {:.4f}...".format(loss.item()),
                      "Test Loss: {:.4f}".format(np.mean(test_losses)))
```

```python
text = get_user_text(0)
encoded = encode(text)
n_hidden = 512
n_layers = 2
net = CharRNN(text, n_hidden, n_layers, drop_prob=0.7)
print(net)
```

```python
train(net, encoded, lr=1e-2, batch_size=32)
```
